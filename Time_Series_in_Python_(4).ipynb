## Step 1: Load data

Read the data into a DataFrame.

```python
import pandas as pd

df_2020 = pd.read_csv('https://nyu.box.com/shared/static/40srcp74ayro33bxls3z78zs6be9k76g.zip', low_memory=False)
df_2020
```

## Step 2: Convert dates

Copy code from [Lecture 4](https://colab.research.google.com/github/afeld/python-public-policy/blob/master/lecture_4.ipynb#scrollTo=CHZ-Pqj0bS9w) to convert the `Created Date` to a `datetime`. Print out the types of the columns using [`.info()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.info.html).

```python
# your code here
df_2020['Created Date'] = pd.to_datetime(df_2020['Created Date'], format='%m/%d/%Y %I:%M:%S %p')
df_2020['Closed Date'] = pd.to_datetime(df_2020['Closed Date'], format='%m/%d/%Y %I:%M:%S %p')
df_2020.info(verbose=True)
df_2020
```

## Step 3: Date counts

Create a DataFrame called `date_counts` that has the count of complaints per Complaint Type per day, then display it.

```python
# your code here
date_counts=df_2020.groupby('Complaint Type').resample('D',on='Created Date').size().reset_index(name='count_requests')
#date_counts=df_2.resample('D',on='Created Date').size().reset_index(name='count_requests')
date_counts
```

## Step 4: Plotting over time

Create a line chart of the count of complaints over time, one line per `Complaint Type`.

```python
# your code here
import plotly.express as px
fig = px.line(date_counts, x="Created Date", y="count_requests", color='Complaint Type')
fig.show()
```

## Step 5: March 2020 counts

Create a DataFrame called `mar_counts` that has the count of each `Complaint Type` in March 2020 in a column called `2020`. Use [`.to_frame()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.to_frame.html) (instead of [`.reset_index()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.reset_index.html)) to use the `Complaint Type` as the index. It should end up looking something like this:

Complaint Type | 2020
--- | ---
APPLIANCE | 806
Abandoned Vehicle | 2468
Air Quality | 643
… | …

_Note there is no numeric index._

```python
# your code here
mar_counts = date_counts[(date_counts['Created Date'].dt.month == 3) & (date_counts['Created Date'].dt.year == 2020)].groupby(by='Complaint Type').sum().rename(columns={'count_requests':'2020'})
mar_counts
#date_counts['2020'] = date_counts.groupby(pd.Grouper(freq='M'))
#df.groupby(pd.Grouper(freq='M'))

#df.groupby(df['date'].dt.strftime('%B'))['Revenue'].sum().sort_values()
```

## Step 7: March 2019 counts

1. Get the `Complaint Type` counts for March 2019.
1. Add these to the `mar_counts` DataFrame as a column called `2019`.

```python
# your code here
mar_2019['datetime'] = pd.to_datetime(mar_2019['Created Date'], format='%m/%d/%Y %I:%M:%S %p')
mar_2019['Closed Date'] = pd.to_datetime(mar_2019['Closed Date'], format='%m/%d/%Y %I:%M:%S %p')
date_counts2019=mar_2019.groupby('Complaint Type').resample('D',on='datetime').size().reset_index(name='count_requests')
mar2019counts = date_counts2019[(date_counts2019['datetime'].dt.month == 3) & (date_counts2019['datetime'].dt.year == 2019)].groupby(by='Complaint Type').sum().rename(columns={'count_requests':'2019'})
mar2019counts
mar_counts = pd.concat([mar_counts,mar2019counts], axis=1, join='inner')
mar_counts
```

## Step 8: Percent change

Use `mar_counts` to calculate the percent change from March 2019 to March 2020 for each `Complaint Type`. Save as the `pct_change` column. Reminder that you can do arithmetic between columns just like you would between variables or even numbers. Should result in something like this:

Complaint Type | 2020 | 2019 | pct_change
--- | --- | --- | ---
APPLIANCE | 806 | 1042 | -0.22
Abandoned Vehicle | 2468 | 1 | 2467.00
Air Quality | 643 | 642 | 0.00
… | … | … | …

```python
# your code here
mar_counts['pct_change']= ((mar_counts['2020']-mar_counts['2019'])/mar_counts['2019'])
mar_counts
```

## Step 9: Filter

Filter to `Complaint Type`s that both:

- Occurred at least 50 times in March 2020
- Changed (increased _or_ decreased) by more than 90%

and save the DataFrame as `top_changed`. A couple of things that may be helpful:

- [Selecting Subsets of Data in Pandas](https://medium.com/dunder-data/selecting-subsets-of-data-in-pandas-39e811c81a0c#0eb4), starting from "Multiple condition expression"
- [Getting absolute values](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.abs.html)

```python
# your code here
top_changed= mar_counts.loc[(mar_counts["2020"]>50) & abs((mar_counts["pct_change"])>0.9)]
top_changed
```

## Step 10: Top changed

Filter the `date_counts` to only the `top_changed` `Complaint Type`s. Save as `top_changed_by_day`.

```python
# your code here
#date_counts=df_2020.groupby('Complaint Type').resample('D',on='Created Date').size().reset_index(name='count_requests')
top_changed_by_day= pd.merge(date_counts, top_changed, on=['Complaint Type'], how='inner')
top_changed_by_day=top_changed_by_day.drop(columns=['2020','2019','pct_change'])
#date_counts[date_counts.set_index(['Complaint Type']).index.isin(top_changed.set_index(['Complaint Type']).index)]
top_changed_by_day
```

## Step 11: Plotting changed complaints

Make a similar plot to Step 4, but with only the top complaints (`top_changed_by_day`).

```python
# your code here
import plotly.express as px
fig = px.line(top_changed_by_day, x="Created Date", y="count_requests", color='Complaint Type')
fig.show()
```

## Bonus: Charting against COVID-19 case counts

Let's take a look at the `Consumer Complaint`s against the COVID-19 case numbers in NYC in the same graph. You'll need to:

1. Find data that provides the COVID-19 case counts for NYC by day.
1. Create a DataFrame with only the `Consumer Complaint` `Complaint Type` counts, by day.
1. Chart the two against each other for February through March.

The result should look something like this:

![bonus solution chart](https://github.com/afeld/python-public-policy/blob/main/img/hw_4_bonus.png?raw=1)

Some resources that may be helpful, minus the blacked-out part:

- [Reading CSV data from GitHub](https://projectosyo.wixsite.com/datadoubleconfirm/single-post/2019/04/15/Reading-csv-data-from-Github---Python)
- [Two Y Axes in plotly](https://plotly.com/python/multiple-axes/#two-y-axes)
   - Note that the `plotly.graph_objects` syntax is a bit different than the `plotly.express` syntax we've been using. With `go.Scatter()`, you don't provide the DataFrame and the names of the columns; you pass `x` and `y` as lists/Series of the values themselves.
- [Setting the Range of Axes Manually in plotly](https://plotly.com/python/axes/#setting-the-range-of-axes-manually)

```python
# your code here
covid19=pd.read_csv('/content/drive/MyDrive/COVID-19_counts.csv', low_memory=False)
covid19
consumercomplaint=top_changed_by_day.loc[top_changed_by_day["Complaint Type"] =="Consumer Complaint"]
consumercomplaint
import plotly.graph_objects as go
from plotly.subplots import make_subplots
fig = go.Figure(go.Scatter(
    mode = "lines+markers",
    y = [1000,2000,3000,4000,5000,6000],
    x = ["Feb 2 2020", "Feb 16", "Mar 1", "Mar 15", "Mar 29"]))
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(
    go.Scatter(covid19,x=["Feb 2 2020", "Feb 16", "Mar 1", "Mar 15", "Mar 29"], y=[1000,2000,3000,4000,5000,6000],name="Case Count"),
    secondary_y=False,
)
fig.add_trace(
    go.Scatter(consumercomplaint,x=["Feb 2 2020", "Feb 16", "Mar 1", "Mar 15", "Mar 29"], y=[1000,2000,3000,4000,5000,6000], name="Consumer Complaints"),
    secondary_y=True,
)
fig.update_layout(
    title_text="Consumer Complaints in 311 vs Covid Cases in NYC"
)
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="<b>primary</b> Case Count", secondary_y=False)
fig.update_yaxes(title_text="<b>secondary</b> Case Count", secondary_y=True)

fig.show()
```
